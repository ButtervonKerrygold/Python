## In-place-Operatoren

### Einführendes Beispiel

❓ Was bewirkt der `+=`-Operator? 

````python
x = 0
x += 1
x += 1
x += 1
print(x)
````

### Weitere In-Place-Operatoren

````python
x = 0
x += 10
x *= 10
x /= 4
x -= 2
print(x)
````

❓ Wie können wir sicherstellen, dass das Ergebnis vom Typ `int` ist?

````python
x = 0
x += 10
x *= 10
x /= 4
x = round(x) # Wie bisher
x -= 2
print(x)
````

💡 **Merke:** Nicht für jede Operation gibt es einen In-Place-Operator.

### Übersicht: In-place-Operatoren

| Kurzschreibweise | Langform     |
| ---------------- | ------------ |
| `x += 1`         | `x = x + 1`  |
| `x -= 1`         | `x = x - 1`  |
| `x *= 2`         | `x = x * 2`  |
| `x /= 2`         | `x = x / 2`  |
| `x //= 2`        | `x = x // 2` |
| `x **= 2`        | `x = x ** 2` |


### ❓ Wie lautet die Ausgabe?

````python
i = 1
summe = 0

while i <= 3:
    summe += i
    i += 1

print(summe)
````

👉 Wir addieren `0 + 1 + 2 + 3`. Ausgabe: `6`.

### In-Place-String-Konkatenation

````python
wort = 'ZIEGEL'
wort = wort + 'DACH'
print(wort)
````

````python
wort = 'KASTEN'
wort = 'BROT' + wort
print(wort)
````

### Konkatenation mit In-Place-Operator

````python
wort = 'SCHLOSS'
wort += 'TÜR'
print(wort)
````

👉 *Ausgabe:* `SCHLOSSTÜR`, nicht `TÜRSCHLOSS`.

### In-Place-Slicing

````python
tag = '<input>'
tag = tag[1:-1]
print(tag)
````

👉 `tag[1:-1]` entfernt das erste und letzte Zeichen eines Strings.

💡 Es handelt sich ebenfall um eine In-Place-Operation ohne eigenen In-Place-Operator.

## Teil 2: `for`-Schleifen

### Meine erste `for`-Schleife

❓ Was passiert hier?

````python
for i in range(4):
    print(i)
````

### `for`-Schleifen sind die *einfachere* Art zu zählen.

- mit `while`-Schleife:
```python
        i = 0
        while i < 4:
            print(i)
            i += 1
```

- mit `for`-Schleife:
```python
        for i in range(4):
            print(i)
```

### Intervallgrenzen

````python
for j in range(3):
    print(j)
````

💭 `range(3)` ist das Folge `0, 1, 2`. 
- Hätten Sie das erwartet? 
- Ist das sinnvoll?

### `for`-Schleife zur reinen Wiederholung

````python
for a in range(5):
    print('Hallo!')
````

💡 Wir sehen der Schleife `for a in range(5):` sofort an, **wie oft** sie ausgeführt wird.

### `for`-Schleife zum Iterieren über Strings

````python
wort = "THU"
for i in range(len(wort)):
    print(wort[i])
````

💡 Die Zahlenfolge `range(len(wort))` entspricht **genau** den Indizes des Strings `wort`. -- Beim Beispiel `"THU"`: `0, 1, 2`.

### `range` mit Startwert

````python
for i in range(2, 5):
    print(i)
````

👉 `range(2, 5)` ist die Folge `2, 3, 4` -- der **Startwert** ist *enthalten*, der **Endwert** *nicht*.

### `range` mit Schrittweite

````python
for i in range(0, 10, 2):
    print(i)
````

👉 `range(0, 10, 2)` ist die Folge `0, 2, 4, 6, 8`. Der dritte Parameter ist die **Schrittweite**.

````python
for i in range(4, 0, -1):
    print(i)
````

👉 `range(4, 0, -1)` ist die Folge `4, 3, 2, 1` -- der Startwert (`4`) ist enthalten, der Endwert (`0`) *nicht*.

### ❓ Warum mussten wir zuerst `while`-Schleifen lernen, wenn `for`-Schleifen doch einfacher sind?

1. `while`-Schleifen schaffen ein Verständnis für Schleifen im Allgemeinen.
2. **Nicht jede Schleife (!)**, lässt sich durch eine `for`-Schleife ersetzen.

### Beispiel: Eingabe validieren

````python
name = ''
while len(name) < 3:
    print('Benutzernamen wählen (min. 3 Zeichen):')
    name = input()
print(f'Neuer Benutzername: {name}')
````

❓ Warum lässt sich dieses Programm nicht mit einer `for`-Schleife schreiben?

### `for` und `while`

- `while`-Schleifen sind die **allgemeinere** Schleifenform.
- `for`-Schleifen sind nur sinnvoll, wenn die **Anzahl der Wiederholungen** im Voraus bekannt ist.
- Dann sind sie aber **deutlich einfacher** zu schreiben und zu lesen.

## Teil 4: Listen

### Meine erste Liste

````python
a = [0, 7, 3, 1]
print(a)
````

💡 Listen sind Folgen von Werten.

👉 Listen werden mit eckigen Klammern `[]` definiert und die Werte durch Kommas getrennt.

### Listenelemente indizieren

````python
a = [0, 7, 3, 1]
print(a[0])
print(a[1])
print(a[2])
````

👉 Listenelemente werden indiziert wie Zeichen in einem String.

### Negative Indizes auf Listen

````python
a = [0, 7, 3, 1]
print(a[-1])
````

👉 Auch auf Listen funktionieren negative Indizes.

### Länge von Listen

````python
a = [0, 7, 3, 1]
print(len(a))
````

👉 Mit `len()` ermitteln wir die Länge einer Liste.

### Die leere Liste

````python
a = [0, 7, 3, 1]
b = []
c = a + b

print(len(a))
print(len(b))
print(len(c))
````

👉 Mit `len()` ermitteln wir die Länge einer Liste.

### Der `+`-Operator auf Listen

````python
a = [1, 0, 0]
b = [0, 1, 0]
c = a + b
print(c)
````

👉 Wir können Listen mit dem `+`-Operator konkatenieren (=verketten).

````python
d = [1]
e = 3 * d
print(e)
````

👉 Listen-Multiplikation funktioniert analog zur String-Multiplikation.

## Teil 5: Listen als Parameter und Rückgabewerte

### ❓ Was tut diese Funktion? Was wäre ein guter Name für die Funktion?

````python
def foo(a: list[int]) -> int:
    s = 0
    for i in range(len(a)):
        s += a[i]
    return s

print(foo([0, 7, 3, 1]))
````

### Listen als Parameter

- Wir können Listen als **Parameter** an Funktionen übergeben.
- Type Hint: `list[<Typ der Elemente>]`
- Beispiele:
    - `list[int]` für Listen von ganzen Zahlen
    - `list[str]` für Listen von Strings
    - `list[float]` für Listen von Gleitkommazahlen

## Teil 6: Zufallszahlen (`int`)

````python
from random import randrange

for i in range(4):
    z = randrange(5)
    print(z)
````

👉 `randrange(5)` gibt einen zufälligen `int`-Wert aus der Menge `{0, 1, 2, 3, 4}` zurück. (Die `0` ist enthalten, `5` ist nicht enthalten.)

````python
from random import randrange

for i in range(4):
    z = randrange(2, 5)
    print(z)
````

👉 `randrange(2, 5)` gibt einen zufälligen `int`-Wert im Bereich `{2, 3, 4}` zurück. (Der **Startwert** `2` ist enthalten, der **Endwert** `5` ist *nicht* enthalten.)
